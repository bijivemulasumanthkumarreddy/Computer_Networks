```python
import matplotlib.pyplot as plt
data='10110111001'
y=[]
for bit in data:
  if bit=='0':
    y+=[0]*10
  else:
    y+=[1]*10
plt.plot(y)
plt.title('NRZ')
plt.ylabel('Voltage')
plt.xlabel('Time')
plt.show()
```

```python
import matplotlib.pyplot as plt
data='101101010001'
y=[1]
for i,bit in enumerate(data):
  if bit=='0':
    y+=[y[i]]*10
  else:
    y+=[int(not y[i])]*10
plt.plot(y)
plt.title('NRZI')
plt.ylabel('Voltage')
plt.xlabel('Time')
plt.show()
```

```python
import socket
HOST = '127.0.0..91'
PORT = 12345
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.bind((HOST, PORT))
server_socket.listen()
print(f"Server listening on {HOST}:{PORT}")
client_socket, client_address = server_socket.accept()
print(f"Connection established with {client_address}")

while True:
    data = client_socket.recv(1024)
    if not data:
        break
    print(f"Received: {data.decode()}")
client_socket.close()
server_socket.close()
```
